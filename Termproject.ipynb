# 라틴아메리카의 IT산업 가능성

1. 프로젝트 목표 및 내용

 스페인어학과로서 중남미 아메리카의 가능성을 높이 평가하고 있다. 중남미 아메리카는 인구수도 많고 성장동력을 충분히 지니고 있다.<br> 
 이에 따라 중남미의 데이터는 점점 많아지는 추세이고 그렇기 때문에 데이터를 활용하면 IT쪽으로 성장가능성이 높다고 판단하고 있었다.<br>
 이번 프로젝트를 하면서 자료를 시각화하여 그러한 가능성을 객관적으로 평가해보려고 한다.  

2. 주제 선정 이유 또는 이 프로젝트의 필요성

 추상적으로 '중남미 아메리카가 발전 가능성이 높다, 블루오션이다' 라는 말을 흔히 들을 수 있다. 그 중 내가 눈여겨 보는 것은 데이터 쪽이다. 인구가 많고 한국에 비해 상대적으로 IT 강국은 아니기 때문에 데이터를 이용하면 IT산업에서 여러 기회를 잡을 수 있을 것 같다고 생각한다. 그러기에 앞서 객관적이고 정확한 데이터가 필요하다. <br>
 우선 데이터에 관한 파일을 불러온 후 한국과 여러 중남미 아메리카를 여러 그래프를 통해 비교해 볼것이다. 또한 각 나라마다 데이터를 따로 저장해 더욱 자세한 정보를 시각화할 것이다. 가능하다면 각 나라마다 불러온 데이터를 비교할 수 있는 그래프 또한 만들어 보고 싶다.

3. 데이터 획득

한국 인터넷 진흥원에서 공개한 '중남미 시장'pdf 파일, hootsuite가 제공하는 'datareportal', 통계청(kosis)에서 제공하는 '범죄발생', 'IMD WORLD DIGITAL COMPETITIVENESS RANKING 2020', 정보통신정책원에서 공개한 'KISDISTAT REPORT', 'ITU의 ICT 발전지수(IDI)를 활용한
정보경제의 성과분석' 를 기반으로 데이터를 획득, 가공하고 내용을 취합했다.
